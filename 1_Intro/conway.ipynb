## Conway's Game of "Life"

Conway's game of "Life" is an excellent illustration of how simple rules can generate complex formations over a time, such as stable patterns and recurring cycles. Let's take a look at how this works in practice:

```python

import conway # custom module for today
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
from IPython.display import Image
```

Using the `randomGrid` function from our custom `conway` module, we can generate a random arrangement of pixels that are "alive" or "dead" ("ON" or "OFF"):

```python
# Set grid size
N = 100

# Construct starting grid with random ON(20%)/OFF(80%)
grid = conway.randomGrid(N)

# Take a look at starting grid:
plt.imshow(grid);
```

If we then update that grid according to Conway's rules (see Gardner 1970), we will see some oscillating patterns emerge over time, as well as some stable patterns.

```python
# set up animation
ani = conway.animate(grid, N)

# Explore Evolution via JavaScript Widget (note stable oscillators):
HTML(ani.to_jshtml())
```

If we take a closer look at some of the specific starting shapes posited by those who have played the game of "Life," such as the "glider," we can see that these behave as expected -- moving in a predictable fashion:

```python
# Make an empty grid
N=10
grid = np.zeros(N * N).reshape(N, N)

# Add a "glider" (Gardner 1970:123)
conway.addGlider(1, 1, grid)

# set up animation
ani = conway.animate(grid, N)

# Explore Evolution via JavaScript Widget:
HTML(ani.to_jshtml())
```

We can also observe stable patterns emerge from this simple set of rules, such as the "Honey Farm":

```python
# Make an empty grid
N=25
grid = np.zeros(N * N).reshape(N, N)
conway.addHoneyFarm(10, 10, grid)

# set up animation
ani = conway.animate(grid, N)

# Explore Evolution via JavaScript Widget:
HTML(ani.to_jshtml())
```

Note that we can save these animations to `gif` format as well if you want to share them outside of Jupyter Notebooks:

```python
# Make an empty grid
N=25
grid = np.zeros(N * N).reshape(N, N)
conway.addHoneyFarm(10, 10, grid)

# set up animation
ani = conway.animate(grid, N)
ani.save('honey_farm.gif',
         writer=animation.PillowWriter(fps=5))

# Open up gif (can then embed in HTML, etc.)
with open('honey_farm.gif','rb') as file:
    display(Image(file.read()))
```

**Takeaways**:
* In the same way that Conway's simple rules can generate complex phenomena over time, so can seemingly simple social and behavioral interactions.
* Simulating phenomena from first principles allows us to observe potentially emergent properties of social systems (e.g. we understand micro-level behavioral patterns, but want to see how their interaction plays itself out at a macro, systems-level)
* We can also try to generate simulated analogs to social phenomena that we empirically observe in the world (e.g. computational theorization of what micro-level behavioral patterns might have produced the macro-level patterns we observe). If we are able to "grow" what we observe, this can help us understand the potential root causes of the phenomena and investigate ways in which we can enact social change.
